## DESCRIPTION :: Student Performance Data Set 

This data approach student achievement in secondary education of two Portuguese schools. The data attributes include student grades, demographic, social and school related features) and it was collected by using school reports and questionnaires. Two datasets are provided regarding the performance in two distinct subjects: Mathematics (mat) and Portuguese language (por). In [Cortez and Silva, 2008], the two datasets were modeled under binary/five-level classification and regression tasks. Important note: the target attribute G3 has a strong correlation with attributes G2 and G1. This occurs because G3 is the final year grade (issued at the 3rd period), while G1 and G2 correspond to the 1st and 2nd period grades. It is more difficult to predict G3 without G2 and G1, but such prediction is much more useful (see paper source for more details).

### So, it looks like it'll be 2 different reports.  But, before I split the reports, I'll clean up the data while I still have just 2 data sets

```python
# SOME ARE SPECIFIC

to_replace_address = {
    'U': 'urban',
    'R': 'rural'
}

to_replace_Pstatus = {
    'T': 'together',
    'A': 'apart'
}

to_replace_famsize = {
    'LE3': 'small',
    'GT3': 'big'
}

to_replace_traveltime = {
    1: 'less than 15 minutes',
    2: '15 to 30 minutes',
    3: '30 minutes to 1 hour',
    4: 'greater than 1 hour',
}

to_replace_studytime = {
    1: 'less than 2 hours',
    2: '2 to 5 hours',
    3: '5 to 10 hours',
    4: 'greater than 10 hours',
}



# SOME ARE STANDARD

to_replace_edu= {
    0: 'none',
    1: 'through 4th grade',
    2: '5th to 9th grade',
    3: 'secondary education',
    4: 'higher education',
}

to_replace_highlow = {
    1: 'very low',
    2: 'low',
    3: 'moderate',
    4: 'high',
    5: 'very high',
}

to_replace_goodbad = {
    1: 'very bad',x
    2: 'bad',
    3: 'ok',
    4: 'good',
    5: 'very good',
}

# SO, THE CHANGES BEING MADE ARE ::

#1 --> leave as is
#2 --> leave as is
#3 --> leave as is
#4 --> CHANGE
math['address'] = math['address'].apply(lambda x:to_replace_address[x])
por['address'] = por['address'].apply(lambda x:to_replace_address[x])
#5 --> CHANGE
math['famsize'] = math['famsize'].apply(lambda x:to_replace_famsize[x])
por['famsize'] = por['famsize'].apply(lambda x:to_replace_famsize[x])
#6 --> CHANGE
math['Pstatus'] = math['Pstatus'].apply(lambda x:to_replace_Pstatus[x])
por['Pstatus'] = por['Pstatus'].apply(lambda x:to_replace_Pstatus[x])
#7 --> CHANGE
math['Medu'] = math['Medu'].apply(lambda x:to_replace_edu[x])
por['Medu'] = por['Medu'].apply(lambda x:to_replace_edu[x])
#8 --> CHANGE
math['Fedu'] = math['Fedu'].apply(lambda x:to_replace_edu[x])
por['Fedu'] = por['Fedu'].apply(lambda x:to_replace_edu[x])
#9 --> leave as is
#10 --> leave as is
#11 --> leave as is
#12 --> leave as is
#13 --> CHANGE
math['traveltime'] = math['traveltime'].apply(lambda x:to_replace_traveltime[x])
por['traveltime'] = por['traveltime'].apply(lambda x:to_replace_traveltime[x])
#14 --> CHANGE
math['studytime'] = math['studytime'].apply(lambda x:to_replace_studytime[x])
por['studytime'] = por['studytime'].apply(lambda x:to_replace_studytime[x])
#15 --> leave as is
#16 --> leave as is
#17 --> leave as is
#18 --> leave as is
#19 --> leave as is
#20 --> leave as is
#21 --> leave as is
#22 --> leave as is
#23 --> leave as is
#24 --> CHANGE
math['famrel'] = math['famrel'].apply(lambda x:to_replace_goodbad[x])
por['famrel'] = por['famrel'].apply(lambda x:to_replace_goodbad[x])
#25 --> CHANGE
math['freetime'] = math['freetime'].apply(lambda x:to_replace_highlow[x])
por['freetime'] = por['freetime'].apply(lambda x:to_replace_highlow[x])
#26 --> CHANGE
math['goout'] = math['goout'].apply(lambda x:to_replace_highlow[x])
por['goout'] = por['goout'].apply(lambda x:to_replace_highlow[x])
#27 --> CHANGE
math['Dalc'] = math['Dalc'].apply(lambda x:to_replace_highlow[x])
por['Dalc'] = por['Dalc'].apply(lambda x:to_replace_highlow[x])
#28 --> CHANGE
math['Walc'] = math['Walc'].apply(lambda x:to_replace_highlow[x])
por['Walc'] = por['Walc'].apply(lambda x:to_replace_highlow[x])
#29 --> CHANGE
math['health'] = math['health'].apply(lambda x:to_replace_goodbad[x])
por['health'] = por['health'].apply(lambda x:to_replace_goodbad[x])
#30 --> leave as is
```
